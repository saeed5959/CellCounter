```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.signal import convolve2d
import os
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.callbacks import TensorBoard

```

```python
#cnn
data_gnrt = ImageDataGenerator(rescale=1/255,rotation_range=180)

valid_gnrt = ImageDataGenerator(rescale=1/255)

data=data_gnrt.flow_from_directory(
    'G:/data/data_',
    target_size=(400,400),
    batch_size=20,
    class_mode="categorical")

valid=valid_gnrt.flow_from_directory(
    'G:/data/valid',
    target_size=(400,400),
    batch_size=5,
    class_mode="categorical")
```

```python
m=models.Sequential()
m.add(layers.Conv2D(16,(3,3),activation='relu',input_shape=(400,400,3)))
m.add(layers.MaxPooling2D((2,2)))
m.add(layers.Conv2D(32,(3,3),activation='relu'))
m.add(layers.MaxPooling2D((2,2))) 
m.add(layers.Conv2D(64,(3,3),activation='relu'))
m.add(layers.MaxPooling2D((2,2)))
m.add(layers.Flatten())
m.add(layers.Dense(512,activation='relu'))
m.add(layers.Dense(5,activation='sigmoid'))

m.compile(loss='categorical_crossentropy',optimizer=RMSprop(lr=0.001),metrics=['acc'])

#tensorboard
tensor=TensorBoard(log_dir='G:/t1', histogram_freq=1)

history=m.fit_generator(
data,
epochs=25,
validation_data=valid,
callbacks=[tensor])

m.save('G:/my_model25.h5')
```

```python
#10 epoch and sigmoid
test_gnrt = ImageDataGenerator(rescale=1/255)

test = test_gnrt.flow_from_directory(
    'G:/k',
    target_size=(400,400),
    batch_size=1,
    class_mode="categorical")

m.predict_classes(test)
```

```python
#w1 = cv2.imread("G:/eso-6346233790.jpg")

test_gnrt = ImageDataGenerator(rescale=1/255)

test = test_gnrt.flow_from_directory(
    'G:/k',
    target_size=(400,400),
    batch_size=1,
    class_mode="categorical")

ans = m5.predict_classes(test)

#local histogram equalizer
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
        

for i in range(0,len(test)):
    if ans[i]==4:
        
        w3 = np.copy(test[i][0][0]*255)
        w3 = np.array(w3,dtype=np.uint8)
        wg = w3[:,:,1]
        
        wg = clahe.apply(wg)
        
        #threshold image
        ret,wg_th = cv2.threshold(wg,170,255,cv2.THRESH_BINARY)
        kernel = np.ones((9,9),np.uint8)
        opening = cv2.morphologyEx(wg_th,cv2.MORPH_OPEN,kernel, iterations = 2)
        
        # Marker labelling
        ret, markers = cv2.connectedComponents(opening)
        # Add one to all labels so that sure background is not 0, but 1
        markers = markers+1

        #markers = watershed(pic, markers, mask=thresh)
        markers = cv2.watershed(w3,markers)
        label = np.copy(markers)
        label = np.array(label,dtype=np.uint8)
        label[label==1] = 0
        
        #connect component
        nb_comp,output,sizes,centroids=cv2.connectedComponentsWithStats(label,connectivity=4)
        #taking away the background
        nb_comp-=1; sizes=sizes[2:,-1]; centroids=centroids[1:,:]
        print(sizes)
        j=0
        p = []
        
       
        #drawing X
        for label in range(0,np.size(sizes)):
            if sizes[label]>300:
                p.append(label)
                j = j+1
        p = np.array(p)
        
        
        
        if j == 2:
            if (sizes[p[0]]*0.8)<sizes[p[1]]<(sizes[p[0]]*1.25):
                ans[i]=1
#       if j == 2:
#           if sizes[1]*0.33<sizes[0]<sizes[1]*3:
#               ans[i]=1
          
ans
```
